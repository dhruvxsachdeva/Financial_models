```python
import numpy as np
import matplotlib.pyplot as plt
```

**Weiner process**
$W_{t+\delta t}= W_t + \epsilon * \sqrt(\delta t)$


```python
T = 1        # total time
N = 1000     # steps
M = 10000    # number of paths
dt = T/N
W=np.zeros((M,N+1))
for i in range(1,N+1):
    W[:, i] = W[:, i-1] + np.sqrt(dt) * np.random.randn(M)
```

A **Wiener process** $W_t$ satisfies:

1. $W_0 = 0$  
2. Independent increments  
3. $W_{t+\Delta t} - W_t \sim \mathcal{N}(0,\Delta t)$

That is, the increment of a Wiener process over a small interval $\Delta t$ is normally distributed with mean $0$ and variance $\Delta t$.

To simulate this using standard normals, take $Z \sim \mathcal{N}(0,1)$ and scale by $\sqrt{\Delta t}$ so that the increment has variance $\Delta t$:
$$
X \;=\; \sqrt{\Delta t}\; Z \quad\Rightarrow\quad \mathrm{Var}(X) = (\sqrt{\Delta t})^2\mathrm{Var}(Z) = \Delta t.
$$

Thus the discrete update used in simulation is
$$
W_{t+\Delta t} \;=\; W_t \;+\; \sqrt{\Delta t}\; Z,
$$
which ensures the increments follow the Wiener process properties and that $\mathrm{Var}(W_t)=t$.


```python
time = np.linspace(0, T, N+1) # starts at 0 till T and creates N+1 equal sections 
plt.figure(figsize=(10,6))

for i in range(100):  # plot 10 sample paths
    plt.plot(time, W[i])

plt.title("Simulated Wiener Process Paths")
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.grid(True)
plt.show()
```
